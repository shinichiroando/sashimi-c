# Subhalo Mass Stripping Perturbation Theory up to Second Order

## Introduction

In modeling the mass evolution of subhalos within a cosmological context, it is essential to accurately represent how subhalos gain mass over time due to processes like accretion and lose mass due to tidal stripping. This note provides the mathematical derivation of the perturbative solution for subhalo mass stripping up to second order, accounting for small deviations in key parameters. By properly handling the perturbation expansion, we ensure accurate modeling of subhalo mass evolution.

---

## 1. Mass Evolution Equation

The mass evolution of a subhalo $m(z)$ as a function of redshift $z$ is governed by the differential equation:

$$
\frac{dm}{dz} = \Phi(z) \cdot m(z) \left( \frac{m(z)}{M_{\text{vir}}(z)} \right)^{\zeta(z)}
$$

Where:

- $\Phi(z) = \dfrac{A(z)}{t_{\text{dyn}}(z) H(z) (1 + z)}$
- $A(z)$: Stripping efficiency function.
- $t_{\text{dyn}}(z)$: Dynamical time at redshift $z$.
- $M_{\text{vir}}(z)$: Virial mass of the host halo at redshift $z$.
- $\zeta(z)$: Exponent characterizing the mass dependence of stripping.
- $H(z)$: Hubble parameter at redshift $z$.

---

## 2. Zeroth-Order Solution

Assuming $\zeta(z)$ is approximately zero (i.e., $\zeta(z) = \zeta_0 \approx 0$), the equation simplifies to:

$$
\frac{dm}{dz} = \Phi(z) \cdot m(z)
$$

This is a separable differential equation. Separating variables:

$$
\frac{dm}{m(z)} = \Phi(z) dz
$$

Integrating both sides from the accretion redshift $z_{\text{a}}$ (where $m = m_{\text{a}}$) to $z$:

$$
\int_{m_{\text{a}}}^{m_0(z)} \frac{1}{m} dm = \int_{z_{\text{a}}}^{z} \Phi(z') dz'
$$

This yields the zeroth-order solution:

$$
\ln m_0(z) - \ln m_{\text{a}} = \int_{z_{\text{a}}}^{z} \Phi(z') dz'
$$

$$
\Rightarrow m_0(z) = m_{\text{a}} \cdot \exp\left( \int_{z_{\text{a}}}^{z} \Phi(z') dz' \right)
$$

This solution represents the exponential growth of subhalo mass due to accretion in the absence of significant stripping ($\zeta_0 \approx 0$).

---

## 3. First-Order Perturbation

To account for the small but non-zero $\zeta(z)$, we introduce a perturbation expansion:

- $\zeta(z) = \zeta_0 + \zeta_1(z)$, where $\zeta_0 \approx 0$.
- Expand $m(z)$ as $m(z) = m_0(z) \cdot e^{\epsilon(z)}$, where $\epsilon(z)$ is a small first-order correction.

### 3.1. Derivation of $\epsilon(z)$

Starting from the differential equation:

$$
\frac{dm}{dz} = \Phi(z) \cdot m(z) \left( \frac{m(z)}{M_{\text{vir}}(z)} \right)^{\zeta(z)}
$$

Substituting $m(z) = m_0(z) e^{\epsilon(z)}$ and $\zeta(z) = \zeta_1(z)$:

$$
\frac{d}{dz} \left[ m_0(z) e^{\epsilon(z)} \right] = \Phi(z) m_0(z) e^{\epsilon(z)} \exp\left[ \zeta_1(z) \ln\left( \frac{m_0(z) e^{\epsilon(z)}}{M_{\text{vir}}(z)} \right) \right]
$$

Simplifying:

1. **Left-Hand Side (LHS):**

$$
\frac{d}{dz} \left[ m_0(z) e^{\epsilon(z)} \right] = m_0'(z) e^{\epsilon(z)} + m_0(z) e^{\epsilon(z)} \epsilon'(z)
$$

2. **Right-Hand Side (RHS):**

$$
\Phi(z) m_0(z) e^{\epsilon(z)} \left[ 1 + \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) \right) \right]
$$

   - Used the approximation $\exp(\zeta_1(z) x) \approx 1 + \zeta_1(z) x$ since $\zeta_1(z)$ is small.

3. **Equate LHS and RHS:**

$$
m_0'(z) e^{\epsilon(z)} + m_0(z) e^{\epsilon(z)} \epsilon'(z) = \Phi(z) m_0(z) e^{\epsilon(z)} \left[ 1 + \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) \right) \right]
$$

4. **Cancel Common Factors:**

Divide both sides by $e^{\epsilon(z)}$:

$$
m_0'(z) + m_0(z) \epsilon'(z) = \Phi(z) m_0(z) \left[ 1 + \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) \right) \right]
$$

5. **Use $m_0'(z) = \Phi(z) m_0(z)$:**

From the zeroth-order solution:

$$
m_0'(z) = \Phi(z) m_0(z)
$$

6. **Subtract $m_0'(z)$ from both sides:**

$$
m_0(z) \epsilon'(z) = \Phi(z) m_0(z) \left[ \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) \right) \right]
$$

7. **Divide by $m_0(z)$:**

$$
\epsilon'(z) = \Phi(z) \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) \right)
$$

Since $\epsilon(z)$ is small (first-order), we can neglect the $\epsilon(z)$ term on the right-hand side (second-order effect):

$$
\epsilon'(z) = \Phi(z) \zeta_1(z) \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right)
$$

**Integrate Both Sides:**

$$
\epsilon(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \ln\left( \frac{m_0(z')}{M_{\text{vir}}(z')} \right) dz'
$$

### 3.2. Simplifying $\epsilon(z)$

Using the expression for $m_0(z')$:

$$
\ln m_0(z') = \ln m_{\text{a}} + \int_{z_{\text{a}}}^{z'} \Phi(u) du
$$

Therefore,

$$
\ln\left( \frac{m_0(z')}{M_{\text{vir}}(z')} \right) = \ln m_{\text{a}} + \int_{z_{\text{a}}}^{z'} \Phi(u) du - \ln M_{\text{vir}}(z')
$$

Substitute back into $\epsilon(z)$:

$$
\epsilon(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \left[ \ln m_{\text{a}} + \int_{z_{\text{a}}}^{z'} \Phi(u) du - \ln M_{\text{vir}}(z') \right] dz'
$$

This expression can be split into three separate integrals:

$$
\epsilon(z) = \epsilon_1(z) + \epsilon_2(z) + \epsilon_3(z)
$$

Where:

- $\epsilon_1(z) = \ln m_{\text{a}} \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') dz'$
- $\epsilon_2(z) = -\int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \ln M_{\text{vir}}(z') dz'$
- $\epsilon_3(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \left( \int_{z_{\text{a}}}^{z'} \Phi(u) du \right) dz'$

---

## 4. Second-Order Perturbation

To include second-order effects, we consider the next-order correction $\delta(z)$, expanding $m(z)$ as:

$$
m(z) = m_0(z) \cdot e^{\epsilon(z) + \delta(z)}
$$

### 4.1. Derivation of $\delta(z)$

Starting from:

$$
\frac{dm}{dz} = \Phi(z) \cdot m(z) \left( \frac{m(z)}{M_{\text{vir}}(z)} \right)^{\zeta(z)}
$$

Substituting $m(z) = m_0(z) e^{\epsilon(z) + \delta(z)}$ and expanding as before:

1. **Equate LHS and RHS:**

$$
m_0'(z) e^{\epsilon(z) + \delta(z)} + m_0(z) e^{\epsilon(z) + \delta(z)} \left( \epsilon'(z) + \delta'(z) \right) = \Phi(z) m_0(z) e^{\epsilon(z) + \delta(z)} \exp\left[ \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) + \delta(z) \right) \right]
$$

2. **Simplify RHS:**

Using the second-order Taylor expansion:

$$
\exp\left[ \zeta_1(z) x \right] \approx 1 + \zeta_1(z) x + \frac{1}{2} \zeta_1(z)^2 x^2
$$

Thus, the RHS becomes:

$$
\Phi(z) m_0(z) e^{\epsilon(z) + \delta(z)} \left[ 1 + \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) + \delta(z) \right) + \frac{1}{2} \zeta_1(z)^2 \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) + \delta(z) \right)^2 \right]
$$

3. **Subtract $m_0'(z) e^{\epsilon(z) + \delta(z)}$ from both sides:**

$$
m_0(z) e^{\epsilon(z) + \delta(z)} \left( \epsilon'(z) + \delta'(z) \right) = \Phi(z) m_0(z) e^{\epsilon(z) + \delta(z)} \left[ \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) + \delta(z) \right) + \frac{1}{2} \zeta_1(z)^2 \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} + \epsilon(z) + \delta(z) \right)^2 \right) \right]
$$

4. **Divide both sides by $m_0(z) e^{\epsilon(z) + \delta(z)}$:**

$$
\epsilon'(z) + \delta'(z) = \Phi(z) \left[ \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) + \delta(z) \right) + \frac{1}{2} \zeta_1(z)^2 \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} + \epsilon(z) + \delta(z) \right)^2 \right) \right]
$$

Since $\delta(z)$ is second-order, we can neglect terms involving $\delta(z)$ on the right-hand side (third-order effect):

$$
\epsilon'(z) + \delta'(z) = \Phi(z) \left[ \zeta_1(z) \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) + \epsilon(z) \right) + \frac{1}{2} \zeta_1(z)^2 \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right)^2 \right) \right]
$$

Subtract the first-order equation:

$$
\epsilon'(z) = \Phi(z) \zeta_1(z) \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right)
$$

From both sides:

$$
\delta'(z) = \Phi(z) \zeta_1(z) \epsilon(z) + \frac{1}{2} \Phi(z) \zeta_1(z)^2 \left( \ln\left( \frac{m_0(z)}{M_{\text{vir}}(z)} \right) \right)^2
$$

### 4.2. Integrating $\delta(z)$

Integrate both sides:

$$
\delta(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \epsilon(z') dz' + \frac{1}{2} \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z')^2 \left( \ln\left( \frac{m_0(z')}{M_{\text{vir}}(z')} \right) \right)^2 dz'
$$

---

## 5. Decomposition of $\delta_2(z)$

The term $\delta_2(z)$ involves the square of the logarithm. By substituting $m_0(z')$, we can expand $\delta_2(z)$ explicitly.

### 5.1. Expanding the Square

We have:

$$
\left( \ln\left( \frac{m_0(z')}{M_{\text{vir}}(z')} \right) \right)^2 = \left( \ln m_{\text{a}} + \int_{z_{\text{a}}}^{z'} \Phi(u) du - \ln M_{\text{vir}}(z') \right)^2
$$

Expanding the square:

$$
\begin{align*}
\left( \ln\left( \frac{m_0(z')}{M_{\text{vir}}(z')} \right) \right)^2 &= \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') + \int_{z_{\text{a}}}^{z'} \Phi(u) du \right)^2 \\
&= \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') \right)^2 + 2 \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') \right) \int_{z_{\text{a}}}^{z'} \Phi(u) du + \left( \int_{z_{\text{a}}}^{z'} \Phi(u) du \right)^2
\end{align*}
$$

### 5.2. Expressing $\delta_2(z)$

Substitute back into $\delta_2(z)$:

$$
\delta_2(z) = \frac{1}{2} \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z')^2 \left[ \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') \right)^2 + 2 \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') \right) \int_{z_{\text{a}}}^{z'} \Phi(u) du + \left( \int_{z_{\text{a}}}^{z'} \Phi(u) du \right)^2 \right] dz'
$$

### 5.3. Splitting $\delta_2(z)$ into Components

We can write $\delta_2(z)$ as the sum of three terms:

1. **First Term:**

   $$
   \delta_{2a}(z) = \frac{1}{2} \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z')^2 \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') \right)^2 dz'
   $$

2. **Second Term:**

   $$
   \delta_{2b}(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z')^2 \left( \ln m_{\text{a}} - \ln M_{\text{vir}}(z') \right) \left( \int_{z_{\text{a}}}^{z'} \Phi(u) du \right) dz'
   $$

3. **Third Term:**

   $$
   \delta_{2c}(z) = \frac{1}{2} \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z')^2 \left( \int_{z_{\text{a}}}^{z'} \Phi(u) du \right)^2 dz'
   $$

---

## 6. Final Expression for $m(z)$

Including up to second-order corrections, the mass $m(z)$ is:

$$
m(z) = m_0(z) \cdot e^{\epsilon(z) + \delta(z)}
$$

Where:

- $m_0(z) = m_{\text{a}} \cdot \exp\left( \int_{z_{\text{a}}}^{z} \Phi(z') dz' \right)$
- $\epsilon(z) = \epsilon_1(z) + \epsilon_2(z) + \epsilon_3(z)$
- $\delta(z) = \delta_1(z) + \delta_{2a}(z) + \delta_{2b}(z) + \delta_{2c}(z)$

With:

- $\epsilon_1(z) = \ln m_{\text{a}} \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') dz'$
- $\epsilon_2(z) = -\int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \ln M_{\text{vir}}(z') dz'$
- $\epsilon_3(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \left( \int_{z_{\text{a}}}^{z'} \Phi(u) du \right) dz'$
- $\delta_1(z) = \int_{z_{\text{a}}}^{z} \Phi(z') \zeta_1(z') \epsilon(z') dz'$

---

## 7. Summary of Steps for Calculation

**Step 1: Compute $m_0(z)$**

- Compute $\Phi(z)$ over the redshift range.
- Integrate $\Phi(z)$ from $z_{\text{a}}$ to $z$ to obtain $\int_{z_{\text{a}}}^{z} \Phi(z') dz'$.
- Calculate $m_0(z)$.

**Step 2: Compute First-Order Correction $\epsilon(z)$**

- Compute $\epsilon_1(z)$, $\epsilon_2(z)$, and $\epsilon_3(z)$ separately.
- Sum to obtain $\epsilon(z)$.

**Step 3: Compute Second-Order Correction $\delta(z)$**

- Compute $\delta_1(z)$ using $\epsilon(z)$.
- Compute $\delta_{2a}(z)$, $\delta_{2b}(z)$, and $\delta_{2c}(z)$ using the expanded expressions.
- Sum to obtain $\delta(z)$.

**Step 4: Compute Final Mass $m(z)$**

- Combine $m_0(z)$, $\epsilon(z)$, and $\delta(z)$ to get $m(z)$.

---

## 8. Important Notes

- **Sign Convention:** The positive sign in the differential equation reflects mass growth due to accretion.
- **Assumption of $\zeta_0 \approx 0$:** Simplifies the equations and is valid when $\zeta(z)$ is small.
- **Validity of Perturbation Expansion:** The perturbative approach is appropriate when the corrections $\epsilon(z)$ and $\delta(z)$ are small compared to one.
- **Neglected Terms:** Higher-order terms (third order and beyond) are neglected as they contribute insignificantly.

## Appendix: N-th Order Corrections

The differential equation for $m(z)$ can be expanded to higher orders to obtain more accurate solutions. Recall that the equation is:

$$
\frac{dm}{dz} = \Phi(z) \cdot m(z) \left( \frac{m(z)}{M_{\text{vir}}(z)} \right)^{\zeta(z)}
$$

or equivalently:

$$
\frac{d\ln m}{dz} = \Phi(z) \left( \frac{m(z)}{M_{\text{vir}}(z)} \right)^{\zeta(z)}
$$

Then we expand $m(z)$ as:

$$
m(z) = m_a \exp\left( \sum_{n=0}^{\infty} \epsilon_n(z) \right)
$$

or equivalently:

$$
\ln m(z) = \ln m_a + \sum_{n=0}^{\infty} \epsilon_n(z)
$$

Where $\epsilon_n(z)$ is the $n$-th order correction to the mass evolution. 

The LHS of the (logarithmic) differential equation can be expanded as a power series in $\epsilon_n(z)$:

$$
\begin{align*}
\text{LHS} &= \frac{d\ln m}{dz} \\
&= \frac{d}{dz} \left( \ln m_a + \sum_{n=0}^{\infty} \epsilon_n(z) \right) \\
&= \sum_{n=0}^{\infty} \frac{d\epsilon_n(z)}{dz}
\end{align*}
$$

The RHS of the (logarithmic) differential equation can be expanded as a power series in $\epsilon_n(z)$:

$$
\begin{align*}
\text{RHS} &= \Phi(z) \left( \frac{m(z)}{M_{\text{vir}}(z)} \right)^{\zeta(z)} \\
&= \Phi(z) \exp\left( \zeta(z) \ln\left( \frac{m(z)}{M_{\text{vir}}(z)} \right) \right) \\
&= \Phi(z) \exp\left( \zeta(z) \left( \ln \left(\frac{m_a}{ M_{\text{vir}}(z)}\right) + \sum_{n=0}^{\infty} \epsilon_n(z) \right) \right) \\
\end{align*}
$$

To simplify the equation, we define $\epsilon'_n(z)$ as $\epsilon'_0 = \epsilon_0 + \ln(m_a/M_{\text{vir}}(z))$ and $\epsilon'_n = \epsilon_n$ for $n > 0$. Then the RHS becomes:

$$
\text{RHS} = \Phi(z) \exp\left( \zeta(z) \sum_{n=0}^{\infty} \epsilon'_n(z) \right)
$$

The differential equation can now be written as:

$$
\sum_{n=0}^{\infty} \frac{d\epsilon_n(z)}{dz} = \Phi(z) \exp\left( \zeta(z) \sum_{n=0}^{\infty} \epsilon'_n(z) \right)
$$

We want to solve this equation order by order, so we need to expand the RHS in terms of the perturbative order.

### Expansion of the RHS

Hereafter we omit the argument $z$ for simplicity. The exponential term in the RHS can be expanded as a power series in $\zeta(z)$:

$$
\begin{align*}
\exp\left( \zeta \sum_{n=0}^{\infty} \epsilon'_n \right) &= \sum_{k=0}^{\infty} \frac{1}{k!} \left( \zeta \sum_{n=0}^{\infty} \epsilon'_n \right)^k \\
&= \sum_{k=0}^{\infty} \frac{1}{k!} \zeta^k \left( \sum_{n=0}^{\infty} \epsilon'_n \right)^k \\
\end{align*}
$$

Using the multinomial theorem, we obtain:

$$
\begin{align*}
\sum_{k=0}^{\infty} \frac{1}{k!} \zeta^k \left( \sum_{n=0}^{\infty} \epsilon'_n \right)^k 
&= \sum_{k=0}^{\infty} \frac{1}{k!} \zeta^k \sum_{\sum_i{n_i} = k} \frac{k!}{n_0! n_1! n_2! \ldots} \prod_{i} {\epsilon'_i}^{n_i} \\
\end{align*}
$$

To obtain specific perturbative order terms, we change the order of summation:

$$
\sum_{\sum_i{n_i}=k} = \sum_{m} \sum_{\substack{\sum_i{n_i}=k \\ \sum_i i \cdot n_i = m}}
$$

where $k = \sum_i n_i$ is the total order of the term and $m = \sum_i i \cdot n_i$ is the perturbative order. Then the RHS can be written as:

$$
\text{RHS} = \Phi \sum_{k=0}^{\infty}\zeta^k\sum_{m=0}^{\infty}  \sum_{\substack{\sum_i{n_i}=k \\ \sum_i i \cdot n_i = m}}\prod_{i} \frac{{\epsilon'_i}^{n_i}}{n_i!}
$$

Here we define $f_{k,m} = \sum_{\substack{\sum_i{n_i}=k \\ \sum_i i \cdot n_i = m}}\prod_i \frac{{\epsilon'_i}^{n_i}}{n_i!}$, so that the RHS can be written as:

$$
\text{RHS} = \Phi \sum_{k=0}^{\infty}\sum_{m=0}^{\infty} \zeta^k  f_{k,m}
$$

Here we note that $f_{0,0} = 1$ (corresponding to $\exp(x) = 1 + \cdots$) and $f_{0,m} = 0$ for $m>0$. In other words,

$$
f_{k,m} = \left(\text{summation of } \prod_i \frac{{\epsilon'_i}^{n_i}}{n_i!} \text{ whose perturbative order is } m \text{ and the total order is } k\right)
$$

### Solving the Differential Equation

The differential equation can now be written as:

$$
\sum_{n=0}^{\infty} \frac{d\epsilon_n}{dz} = \Phi \sum_{k=0}^{\infty}\sum_{m=0}^{\infty} \zeta^k  f_{k,m}
$$

To extract the $n$-th order term of the RHS, we change the order of summation:

$$
\sum_{k=0}^{\infty}\sum_{m=0}^{\infty}\zeta^k  f_{k,m} = \Phi \sum_{n=0}^{\infty} \sum_{m=0}^{n} \zeta^{n-m} f_{n-m,m}
$$

where $n = k + m$. Eventually, taking the $n$-th order term of each side, we obtain the $n$-th order correction to the mass evolution equation:

$$
\begin{align*}
\frac{d\epsilon_n}{dz} 
&= \Phi\sum_{m=0}^{n} \zeta^{n-m} f_{n-m,m}
&= \Phi\left(\zeta^n f_{n,0} + \zeta^{n-1} f_{n-1,1} + \zeta^{n-2} f_{n-2,2} + \cdots + f_{0,n}\right)
\end{align*}
$$

Again, we remind that $f_{0,0} = 1$ and $f_{0,m} = 0$ for $m>0$. Moreover, $f_{n,m}$ depends on terms whose perturbative order is less than or equal to $m$, so that the $n$-th order correction depends on lower-order corrections. Therefore, we can solve the differential equation order by order, starting from the zeroth order.

### Example

We check the first few orders of the perturbative expansion. 

#### The zeroth order equation

$$
\frac{d\epsilon_0}{dz} = \Phi f_{0,0} = \Phi
$$

Then we have:

$$
\epsilon_0(z) = \int_{z_a}^{z} \Phi(z') dz'
$$

#### The first order equation

$$
\begin{align*}
\frac{d\epsilon_1}{dz} 
&= \Phi \left( \zeta f_{1,0} + \zeta^0 f_{0,1} \right) \\
&= \Phi \zeta f_{1,0} \\
&= \Phi \zeta \frac{\epsilon'_0}{1!}
\end{align*}
$$

Then we have:

$$
\epsilon_1(z) = \int_{z_a}^{z} \Phi(z') \zeta(z') \epsilon'_0(z') dz'
$$

#### The second order equation

$$
\begin{align*}
\frac{d\epsilon_2}{dz}
&= \Phi \left( \zeta^2 f_{2,0} + \zeta f_{1,1} + \zeta^0 f_{0,2} \right) \\
&= \Phi \zeta^2 f_{2,0} + \Phi \zeta f_{1,1} \\
&= \Phi \zeta^2 \frac{{\epsilon'_0}^2}{2!} + \Phi \zeta \frac{\epsilon'_1}{1!}
\end{align*}
$$

Then we have:

$$
\epsilon_2(z) = \int_{z_a}^{z} \Phi(z') \left( \zeta(z')^2 \frac{\epsilon'_0(z')^2}{2!} + \zeta(z') \epsilon'_1(z') \right) dz'
$$

#### The third order equation

$$
\begin{align*}
\frac{d\epsilon_3}{dz}
&= \Phi \left( \zeta^3 f_{3,0} + \zeta^2 f_{2,1} + \zeta f_{1,2} + \zeta^0 f_{0,3} \right) \\
&= \Phi \zeta^3 f_{3,0} + \Phi \zeta^2 f_{2,1} + \Phi \zeta f_{1,2} \\
&= \Phi \zeta^3 \frac{{\epsilon'_0}^3}{3!} + \Phi \zeta^2 \frac{\epsilon'_0 {\epsilon'_1}}{1!1!} + \Phi \zeta \frac{{\epsilon'_2}}{1!}
\end{align*}
$$

Then we have:

$$
\epsilon_3(z) = \int_{z_a}^{z} \Phi(z') \left( \zeta(z')^3 \frac{{\epsilon'_0(z')}^3}{3!} + \zeta(z')^2 \frac{\epsilon'_0(z') \epsilon'_1(z')}{1!1!} + \zeta(z') \epsilon'_2(z') \right) dz'
$$


This process can be repeated for higher orders to obtain the full perturbative expansion of the mass evolution equation.

```python
import sashimi_c
import matplotlib.pyplot as plt
import numpy as np
import tqdm
import scipy
from scipy.integrate import odeint, cumulative_trapezoid

# # if scipy has cumulative_simpson, load it
if hasattr(scipy.integrate, 'cumulative_simpson'):
    from scipy.integrate import cumulative_simpson
    print("Using cumulative_simpson")
    def _cumulative_simpson(y, x, *args, **kwargs):
        # NOTE: cumulative_simpson works only for increasing x, like F(x) = \int_a^x f(x') dx'.
        # However, we want to obtain \int_a^x f(x') dx' for decreasing x.
        # To do this, we rephrase the integral as follows:
        #   \int_a^x = \int_b^x - \int_b^a, where b < x < a.
        ret = cumulative_simpson(y[::-1], x=x[::-1], *args, **kwargs)[::-1]
        return ret - ret[0]
    # cumulative_simpson = _cumulative_simpson
    # cumulative_integrate = cumulative_simpson
else:
    print("Using cumulative_trapezoid")
    # cumulative_integrate = cumulative_trapezoid

# cumulative_integrate = cumulative_trapezoid


class MySubhaloProperties(sashimi_c.subhalo_properties):
    def __init__(self):
        super().__init__()

    def set_integration_method(self, method):
        self.cumulative_integrate = {
            "cumulative_trapezoid" : cumulative_trapezoid,
            "cumulative_simpson" : _cumulative_simpson
        }[method]

    def Mzvir(self,z):
        Mz200 = self.Mzzi(self.M0,z,0.)
        Mvir = self.Mvir_from_M200_fit(Mz200,z)
        return Mvir

    def AMz(self,z):
        log10a = (-0.0003*np.log10(self.Mzvir(z)/self.Msun)+0.02)*z \
                        +(0.011*np.log10(self.Mzvir(z)/self.Msun)-0.354)
        return 10.**log10a

    def zetaMz(self,z):
        return (0.00012*np.log10(self.Mzvir(z)/self.Msun)-0.0033)*z \
                    +(-0.0011*np.log10(self.Mzvir(z)/self.Msun)+0.026)

    def tdynz(self,z):
        Oz_z = self.OmegaM*(1.+z)**3/self.g(z)
        return 1.628/self.h*(self.Delc(Oz_z-1.)/178.0)**-0.5/(self.Hubble(z)/self.H0)*1.e9*self.yr

    def msolve(self,m, z):
        return self.AMz(z)*(m/self.tdynz(z))*(m/self.Mzvir(z))**self.zetaMz(z)/(self.Hubble(z)*(1+z))

    def subhalo_mass_stripped(self,ma,za,z,n_z=100):
        zcalc = np.linspace(za,z,n_z)
        return zcalc,odeint(self.msolve,ma,zcalc)

    def Phi(self,z):
        """ subhalo stripping factor assuming zetaMz(z) = 0.
        The stripping rate dm/dt is given by
          dm/dt(z) = m(z) * Phi(z) * (m(z)/Mzvir(z))**zetaMz(z)
        """
        return self.AMz(z)/self.tdynz(z)/self.Hubble(z)/(1+z)

    def subhalo_mass_stripped_pert0(self,ma,za,z,n_z=100):
        """ calculate subhalo mass stripping using perturbative method.
        Here we assume zetaMz is almost 0.
        """
        _z = np.linspace(za,z,n_z)
        integrand = self.Phi(_z)
        int_factor = self.cumulative_integrate(integrand,x=_z,initial=0)
        return _z, ma*np.exp(int_factor)


    def _eps_1(self,za,z,n_z=100):
        """ calculate epsilon1 """
        _z = np.linspace(za,z,n_z)
        Phi_z = self.Phi(_z)
        zeta1_z = self.zetaMz(_z)
        integrand = Phi_z * zeta1_z
        return _z, self.cumulative_integrate(integrand,x=_z,initial=0)

    def _eps_2(self,za,z,n_z=100):
        """ calculate epsilon2 """
        _z = np.linspace(za,z,n_z)
        Phi_z = self.Phi(_z)
        zeta1_z = self.zetaMz(_z)
        ln_Mvir_z = np.log(self.Mzvir(_z))
        integrand = - Phi_z * zeta1_z * ln_Mvir_z
        return _z, self.cumulative_integrate(integrand,x=_z,initial=0)

    def _eps_3(self,za,z,n_z=100):
        """ calculate epsilon3 """
        _z = np.linspace(za,z,n_z)
        Phi_z = self.Phi(_z)
        zeta1_z = self.zetaMz(_z)
        int_Phi = self.cumulative_integrate(Phi_z,x=_z,initial=0)
        integrand = Phi_z * zeta1_z * int_Phi
        return _z, self.cumulative_integrate(integrand,x=_z,initial=0)

    def subhalo_mass_stripped_pert1(self, ma, za, z, n_z=100):
        """ calculate subhalo mass stripping using perturbative method.
        The subhalo mass can be written as
        m(z) = m0(z) * exp(epsilon(z))
        where m0(z) is the zeroth-order mass and epsilon(z) is the perturbative correction, given by
        eplison(z) = epsilon1' + epsilon2 + epsilon3
          = (\ln{m_a} * epsilon1 + epsilon2 + epsilon3
        where
        - \epsilon_1 = \int_{z_a}^{z} \Phi(z{\prime}) \cdot \zeta_1(z{\prime}) dz{\prime}
        - \epsilon_2 = - \int_{z_a}^{z} \Phi(z{\prime}) \cdot \zeta_1(z{\prime}) \cdot \ln M_{\text{vir}}(z{\prime}) dz{\prime}
        - \epsilon_3 = \int_{z_a}^{z} \Phi(z{\prime}) \cdot \zeta_1(z{\prime}) \left( \int_{z_a}}^{z{\prime}} \Phi(u) du \right) dz{\prime}
        """
        _z,m0 = self.subhalo_mass_stripped_pert0(ma,za,z,n_z)
        _z,eps1 = self._eps_1(za,z,n_z)
        _z,eps2 = self._eps_2(za,z,n_z)
        _z,eps3 = self._eps_3(za,z,n_z)
        return _z, m0*np.exp(np.log(ma)*eps1+eps2+eps3)


    def subhalo_mass_stripped_pert2(self, ma, za, z, n_z=100):
        """Calculate subhalo mass stripping including second-order perturbation."""
        # Create redshift array
        _z = np.linspace(za, z, n_z)

        # Step 1: Compute Phi(z), zeta1(z), and m0(z)
        Phi_z = self.Phi(_z)
        zeta1_z = self.zetaMz(_z)  # Since zeta0 â‰ˆ 0
        int_Phi = self.cumulative_integrate(Phi_z, x=_z, initial=0)
        m0_z = ma * np.exp(int_Phi)

        # Step 2: Compute ln(m0(z)/Mvir(z))
        ln_m0_over_Mvir = np.log(m0_z / self.Mzvir(_z))

        # Step 3: Compute epsilon(z) (First-order correction)
        _z, eps1 = self._eps_1(za, z, n_z)
        _z, eps2 = self._eps_2(za, z, n_z)
        _z, eps3 = self._eps_3(za, z, n_z)
        epsilon_total = np.log(ma) * eps1 + eps2 + eps3

        # Step 4: Compute delta(z)
        # Compute delta_1(z)
        integrand_delta1 = Phi_z * zeta1_z * epsilon_total
        delta1 = self.cumulative_integrate(integrand_delta1, x=_z, initial=0)

        # Compute delta_2(z)
        integrand_delta2 = 0.5 * Phi_z * zeta1_z**2 * ln_m0_over_Mvir**2
        delta2 = self.cumulative_integrate(integrand_delta2, x=_z, initial=0)

        # Total delta(z)
        delta_z = delta1 + delta2

        # Step 5: Compute the final mass m(z) including second-order correction
        m_z = m0_z * np.exp(epsilon_total + delta_z)

        return _z, m_z

sp = MySubhaloProperties()
sp.M0 = 1.e12*sp.Msun
```
